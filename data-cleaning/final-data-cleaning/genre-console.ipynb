# Explore Genre and Console control variables

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../datasets/final-df-count-models.csv')
```

```python
df.head()
```

```python
genre_counts = df.genre.value_counts()
```

```python
genre_counts
```

```python
genre_counts/sum(genre_counts)
```

```python
df['is_action'] = df['genre'].apply(lambda genre: 1 if genre=='Action' else 0)
```

```python
df.head(10)
```

```python
console_counts = df.console.value_counts()
```

```python
console_counts
```

```python
console_counts/sum(console_counts)
```

```python
df['is_pc'] = df['console'].apply(lambda console: 1 if console=='pc' else 0)
```

```python
df.head()
```

```python
df.to_csv('../../datasets/final-df-count-models.csv', index=False)
```
